## Session 3
## Arima ÌôúÏö© ÏãúÍ≥ÑÏó¥ ÏòàÏ∏°  
* AR(Autoregression) + MA(Moving Average)


### Hyper parameterÌäúÎãù
---

````python
# Define the p, d and 1 parameters to take any balue between 0 and 2
p = d = q = range(0,2)

# Generate all different combinations of p, 1 and 1 triplets
import itertools
pdq = list(itertools.product(p, d, q))

# Generate all different combinations of seasonal p, d and q triplets
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]

print('Examples of parameter combinations for Seasonal ARIMA...')
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[1]))
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[2]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[3]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[4]))
````

````python
# ÌïÑÏöî ÏóÜÎäî Ïª¨Îüº ÏÇ¨Ï†Ñ Ï†úÍ±∞
df = df.drop(['year', 'month'], axis=1)
df.head()
````

````python
# Î™®Îì† Ï°∞Ìï©Ïùò Parameter Î™®Îç∏ ÏÉùÏÑ± ÌõÑ AICÎ°ú Î™®Îç∏ ÏÑ±Îä• ÌååÏïÖ (AICÍ∞Ä ÏûëÏùÑÏàòÎ°ù ÏµúÏ†ÅÏùò Î™®Îç∏Ïùº ÌôïÎ•†Ïù¥ ÎÜíÎã§.)
# 1ÎÖÑ Îã®ÏúÑ Îç∞Ïù¥ÌÑ∞Ïù¥Í∏∞ ÎïåÎ¨∏Ïóê Ï£ºÍ∏∞Í∞Ä 12
import statsmodels.api as sm
import numpy as np
np.random.seed(1234)
warnings.filterwarnings("ignore") # specify to ignore warning messages

cols = ['param', 'paran_seasonal', 'aic']
lst=[]
for param in pdq:
  for param_seasonal in seasonal_pdq:
    try:
      mod = sm.tsa.statespace.SARIMAX(df,
                                      order=param,
                                      seasonal_order=param_seasonal,
                                      enforce_stationarity=False,
                                      enforce_invertibility=False)
      results=mod.fit()

      print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
      lst.append([param, param_seasonal, results.aic])
      df1 = pd.DataFrame(lst, columns=cols)
    except:
      continue
````

````python
df1
````

````python
# AICÍ∞Ä ÏµúÏÜåÍ∞Ä ÎêòÎäî parameter Ï°∞Ìï©
aic_min = df1['aic'].min()
optimal_para=df1.loc[df1['aic']==aic_min]
optimal_para
````

### Î™®Îç∏ ÌïôÏäµ Î∞è ÌèâÍ∞Ä
---
* Î™®Îç∏ÏùÑ ÌïôÏäµÌïòÍ≥† ÏòàÏ∏°ÏùÑ ÏàòÌñâÌïòÏó¨ ÏÑ±Îä•ÏùÑ ÌèâÍ∞Ä

````python
# Best score Í∏∞Ï§Ä Î™®Îç∏ ÌïôÏäµ
mod = sm.tsa.statespace.SARIMAX(df,
                                order=(0,1,1),
                                seasonal_order=(0,1,1,12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)
results = mod.fit()
print(results.summary().tables[1])
````

````python
# Ljung-Box : ÏûîÏ∞®Í∞Ä White noiseÎ•º Îî∞Î•¥ÎäîÏßÄ (ÏãúÍ≥ÑÏó¥ Î™®ÌòïÏù¥ Ïûò Ï†ÅÌï©ÎêòÏóàÍ≥† ÎÇ®ÏùÄ ÏûîÏ∞®Îäî ÎçîÏù¥ÏÉÅ ÏûêÍ∏∞ÏÉÅÍ¥ÄÏùÑ Í∞ÄÏßÄÏßÄ ÏïäÎäî Î∞±ÏÉâ Ïû°Ïùå)
# Jarque-Bera : ÏûîÏ∞®Ïùò Î∂ÑÌè¨Í∞Ä normal distributionÏùÑ Îî∞Î•¥ÎäîÏßÄ
# Hereroskeddasticity : ÏãúÍ∞ÑÎåÄÎ≥Ñ ÏûîÏ∞®Ïùò Î∂ÑÏÇ∞Ïù¥ ÏùºÏ†ïÌïúÏßÄ
# ==>  Î™®Îëê 0.05 Ïù¥ÏÉÅÏù¥ÎØÄÎ°ú White noiseÎ•º Îî∞Î•¥Î©∞, normal distributionÏù¥Î©∞, Î∂ÑÏÇ∞Ïù¥ ÏùºÏ†ïÌï®.
results.summary()
````

````python
# ÏÉùÏÑ±Îêú Î™®Îç∏ Ï†ïÌï©ÏÑ± ÌôïÏù∏ Í≥ºÏ†ï(Arima Î™®Îç∏ÏùÑ ÏÇ¨Ïö©Ìï† Ïàò ÏûàÎäîÏßÄ ÌåêÎã®ÌïòÎäî Í≤É, ÏûîÏ∞®Í∞Ä Ï†ïÏÉÅÏÑ±ÏùÑ Îî∞Î•¥ÎäîÏßÄ)
results.plot_diagnostics(figsize=(15,12))
plt.show()
# (1) Stadardized residual : ÏûîÏ∞®Í∞Ä Î∞±ÏÉâÏû°Ïùå(White noise)Ïóê Í∞ÄÍπåÏö∏ÏàòÎ°ù Ï¢ãÏùå, ÌèâÍ∑† 0ÏùÑ Ï§ëÏã¨ÏúºÎ°ú Î∞±ÏÉâÏû°ÏùåÏùò ÌòïÌÉú
# (2) Histogram plus estimated density : ÏûîÏ∞®Ïùò Î∂ÑÌè¨Í∞Ä Ï†ïÍ∑úÎ∂ÑÌè¨Î•º Îî∞Î•¥ÎäîÏßÄ, ÏûîÏ∞®Ïùò ÌûàÏä§ÌÜ†Í∑∏Îû®Í≥º KDE plotÏùÑ Í∑∏Î¶¨Í≥†, N(0,1) Ï†ïÍ∑úÎ∂ÑÌè¨ÏôÄ ÎπÑÍµê
# (3) Q-Q plot : ÎëêÍ∞úÏùò Î∂ÑÌè¨Í∞Ä Í∞ôÏùÄÏßÄ Îã§Î•∏ÏßÄ ÌåêÎã®ÌïòÎäî Plot(xÏ∂ï : Ïù¥Î°†Ï†Å ÏÉòÌîåÎ∂ÑÌè¨(Ï†ïÍ∑úÎ∂ÑÌè¨), yÏ∂ï: df['M3])
# (4) Correlogram(ÏΩúÎ†êÎ°úÍ∑∏Îû®) : ÏûîÏ∞®Ïóê ÎåÄÌïú ACF, ÏòàÏ∏°Îêú ÏûêÍ∏∞ÏÉÅÍ¥ÄÏÑ±ÏùÑ ÎÇòÌÉÄÎÇ¥Í∏∞ ÏúÑÌïú plot, Ï¥àÎ°ùÏÉâ Box ÏïàÏóê Îì§Ïñ¥ÏôÄ ÏûàÏúºÎ©¥ ÏûêÍ∏∞ÏÉÅÍ¥ÄÏÑ±Ïù¥ ÏóÜÎã§Í≥† ÌåêÎã®Ìï† Ïàò ÏûàÏùå

# ÏïΩÍ∞ÑÏùò Ï†ïÍ∑úÏÑ±ÏùÑ Î≤óÏñ¥ÎÇúÎã§. ÏòàÏ∏° ÌõÑ Ïûò ÎßûÏßÄ ÏïäÏúºÎ©¥ ÌååÎùºÎØ∏ÌÑ∞ ÏàòÏ†ï ÌõÑ Ïû¨ÏßÑÌñâ.
````

### ÏòàÏ∏° Î∞è Plotting


````python
#################################################################################################################
#--------------- start=pd.to_datetime(        ) ----------------------------------------------------------------#
#--------------- ÏòàÏ∏°ÏùÑ ÏãúÏûë Î∞è Ï¢ÖÎ£åÌïòÍ≥† Ïã∂ÏùÄ ÎÇ†ÏßúÎ•º ÎπàÏπ∏Ïóê ÏûÖÎ†•Ìï® ---------------------------------------------#
#--------------- ex) 2017-01-01 --------------------------------------------------------------------------------#
pred = results.get_prediction(start=pd.to_datetime('2017-01-01'), end=pd.to_datetime('2019-10-01'), dynamic=False)
#################################################################################################################

pred_ci = pred.conf_int()
# Í∑∏ÎûòÌîÑÎ•º 2015ÎÖÑ Î∂ÄÌÑ∞ Í∑∏Î¶¨Í≤†Îã§.
ax = df['2012':].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7)

ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='red', alpha=.2)

ax.set_xlabel('Date')
ax.set_ylabel('M3')
plt.legend()

plt.rcParams["figure.figsize"] = (30, 5)
plt.show()
````

````python
y_forecasted = pred.predicted_mean
########################################
# ÏúÑÏóêÏÑú ÏÑ†Ï†ïÌïú ÏòàÏ∏° ÏãúÏûëÎÇ†ÏßúÎ•º ÏûÖÎ†•   #
y_truth = df['2017-01-01':]
########################################

# Compute the mean square error
mse = ((y_forecasted - y_truth['M3']) ** 2).mean()
print('The Mean Squared Error of our forecasts is {}'.format(round(mse, 2)))

mae = (abs(y_forecasted - y_truth['M3'])).mean()
print('The MAE is {}'.format(round(mae, 2)))
````
